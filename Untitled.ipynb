```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dftitlebasics = pd.read_csv('title.basics.tsv.gz', sep='\t', low_memory=False)
```

```python
dftitlebasics['titleType'].value_counts()
```

```python
dftitlebasics['startYear'][dftitlebasics['startYear'] == '\\N'] = -1
dftitlebasics['isAdult'][dftitlebasics['isAdult'] == '\\N'] = 0
# Only use movies will reduce the size drastically
moviecondition = dftitlebasics['titleType'] == 'movie'
isAdultFalse = dftitlebasics['isAdult'].astype(int) == 0
# Only movies newer than 1950
movie_age_ondition = dftitlebasics['startYear'].astype(int) > 1950
dftitle_only_movies = dftitlebasics[moviecondition & movie_age_ondition & isAdultFalse]

dftitle_only_movies.head()
```

```python
del dftitlebasics
dftitle_only_movies.info()
```

```python
dfnamebasics = pd.read_csv('name.basics.tsv.gz', sep='\t') # , nrows=1000)
```

```python
dfratings = pd.read_csv('title.ratings.tsv.gz', sep='\t')
```

```python
dftitleprincipals = pd.read_csv('title.principals.tsv.gz', sep='\t') # , nrows=10000)
```

```python
actorCondition = dftitleprincipals['category'].str.contains('actor|actress', regex=True)
dftitlePrincipalActor = dftitleprincipals[actorCondition]
dftitlePrincipalActor.sample(15)
```

```python
dfActorInMovies = dftitle_only_movies.merge(dftitlePrincipalActor, on='tconst')
```

```python
dfActorInMovies.info()
```

```python
dfActorInMoviesWithNames = dfActorInMovies.merge(dfnamebasics, on='nconst')
```

```python
dfActorInMoviesWithNames.head()
```

```python
dfActorInMoviesWithNames.info()
```

```python
# Join the ratings
dfActorInMoviesWithNamesAndRating = dfActorInMovies.merge(dfratings, on='tconst')
```

```python
dfActorInMoviesWithNamesAndRating.info()
```

```python
cols = dfActorInMoviesWithNamesAndRating.columns
use_cols = cols.drop(['originalTitle', 'isAdult', 'originalTitle', 'isAdult', 'ordering', 'category',
       'job', 'characters', 'averageRating', 'knownForTitles'])
print(use_cols)

dfUseForGraph = dfActorInMoviesWithNamesAndRating[use_cols]

dfUseForGraph.head()
```
